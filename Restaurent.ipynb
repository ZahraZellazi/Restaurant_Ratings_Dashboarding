```python
# Merge ratings_data with restaurants_df based on 'Restaurant_ID'
merged_data = pd.merge(ratings_data, restaurants_df, on='Restaurant_ID', how='inner')

# Group by restaurant 'Name' to get the average ratings
ratings_by_restaurant = merged_data.groupby('Name').agg({
    'Overall_Rating': 'mean',
    'Food_Rating': 'mean',
    'Service_Rating': 'mean'
}).reset_index()

# Visualization 1: Bar plot of average overall rating by restaurant
plt.figure(figsize=(10, 6))
sns.barplot(x='Name', y='Overall_Rating', data=ratings_by_restaurant, palette='coolwarm')
plt.title('Average Overall Rating by Restaurant')
plt.xlabel('Restaurant')
plt.ylabel('Average Overall Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


```

```python
# Visualization 2: Heatmap of correlation between different rating types
plt.figure(figsize=(8, 6))
corr_matrix = merged_data[['Overall_Rating', 'Food_Rating', 'Service_Rating']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Ratings')
plt.tight_layout()
plt.show()

```

```python
# Visualization 3: Box plot for distribution of ratings
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_data[['Overall_Rating', 'Food_Rating', 'Service_Rating']], palette='coolwarm')
plt.title('Distribution of Overall, Food, and Service Ratings')
plt.xlabel('Rating Type')
plt.ylabel('Rating Value')
plt.xticks([0, 1, 2], ['Overall Rating', 'Food Rating', 'Service Rating'])
plt.tight_layout()
plt.show()
```
